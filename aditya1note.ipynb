```
s = "abcABC123"
u = 0
l = 0
d =0
for i in s:
  if i.isupper():
    u +=1
  elif i.islower():
    l +=1
  else:
    d+=1
print("uppercase:", u, "lowercase:", l, "digits:",d)
```
